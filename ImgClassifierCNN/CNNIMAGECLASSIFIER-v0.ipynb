# DATA PROCESSING

### Remove Dodgy Images

```python
# Variable to hold the path to our data directory
data_dir = os.path.join(main_folder_path,'Data for Face Recognition')
```

```python
#Create list of extentions
image_exts = ['jpeg','jpg','bmp','png']
```

```python
image_exts
```

```python
os.listdir(data_dir)
```

```python
for image_class in os.listdir(data_dir):
   print(image_class)
```

```python
for image_class in os.listdir(data_dir):
    for image in os.listdir(os.path.join(data_dir, image_class)):
        image_path = os.path.join(data_dir, image_class, image)
        try:
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in image_exts:
                print('Image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e:
            print('Issue with image {}'.format(image_path))
            #os.remove(image_path)
```

### Load Data

# Preprocess & Scale Data

```python
data = data.map(lambda x, y:(x/255, y))
```

```python
scaled_iterator = data.as_numpy_iterator()
```

```python
batch = scaled_iterator.next()
```

```python
batch[0].max()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img)
    ax[idx].title.set_text(batch[1][idx])
```

# Split Data

```python
len(data)
```

```python
train_size = int(len(data)*.7) # (during) train deep learning model
val_size = int(len(data)*.2)+1 # (during) validation data is evaluating our model while we are training
test_size = int(len(data)*.1)+1# post training
```

```python
train_size
```

```python
val_size
```

```python
test_size
```

```python
train_size+val_size+test_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

# Build Deep Learning Model

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

# Train

```python
logdir= 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

### CHECK MODEL METRICS

#### LOSS

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='magenta', label='val_loss')
fig.suptitle('Loss', fontsize=20, color='blue')
plt.legend(loc="upper left")
plt.show()
```

#### ACCURACY

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='magenta', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20, color='blue')
plt.legend(loc="upper left")
plt.show()
```

# Evaluation of Performance

```python
pre = Precision()
re = Recall()
biacc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    biacc.update_state(y, yhat)
```

```python
print(f'Precision:{pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{biacc.result().numpy()}')
```
